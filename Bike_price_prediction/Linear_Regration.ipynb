```python
from sklearn.model_selection import train_test_split   #scikit-learn
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)  # 80% training size
```

```python
from sklearn.linear_model import LinearRegression  #Algorithm import
```

```python
lnr = LinearRegression()
```

```python
lnr.fit(x_train,y_train)
```

```python
x_train
```

```python
lnr.score(x_train,y_train)*100
```

```python
# unseen data or test data
lnr.score(x_test,y_test)*100
```
